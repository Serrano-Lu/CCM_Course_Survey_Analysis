```python
# Rename columns to lowercase and replace spaces with underscores
df_2023.columns = df_2023.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Confirm the changes by displaying the new column names
df_2023.columns
```

```python
# Clean and condense 'Race/Ethnicity' column
race_mapping = {
    'White/Caucasian': 'White',
    'Black/African American': 'Black',
    'Hispanic or Latino': 'Hispanic',
    'Asian': 'Asian',
    'Multi-Racial': 'Multiracial',
    'Hispanic or Latino;White/Caucasian': 'Multiracial',
}
```

```python
# Apply the mapping to the 'Race/Ethnicity' column
df_2023['race/ethnicity'] = df_2023['race/ethnicity'].replace(race_mapping)
```

```python
# Verify the changes
df_2023['race/ethnicity'].unique()
```

```python
# Clean and condense 'Major' column 
major_mapping = {
    # Engineering majors
    'Mechanical Engineering': 'Engineering',
    'Electrical Engineering': 'Engineering',
    'Civil Engineering': 'Engineering',
    'Chemical Engineering': 'Engineering',
    'Biomedical Engineering': 'Engineering',
    
    # Computing majors
    'Computer Science': 'Computing',
    'Information Technology': 'Computing',
    'Software Development': 'Computing',
    'Cybersecurity': 'Computing',
    
    # Liberal Arts majors
    'Liberal Arts and Sciences': 'Liberal Arts',
    'Liberal Arts': 'Liberal Arts',
    'Humanities': 'Liberal Arts',
    'Social Sciences': 'Liberal Arts',
    
    # Health Sciences majors
    'Nursing': 'Health Sciences',
    'Radiography': 'Health Sciences',
    'Public Health': 'Health Sciences',
    
    # Business majors
    'Business Administration': 'Business',
    'Marketing': 'Business',
    'Finance': 'Business',
    'Accounting': 'Business',
    
    # Musical Arts majors
    'Music Performance': 'Musical Arts',
    'Musical Theatre': 'Musical Arts',
    'Music Technology': 'Musical Arts',
    
    # Psychology majors
    'Psychology': 'Psychology',
    'Counseling Psychology': 'Psychology',
    'Clinical Psychology': 'Psychology',
    'psychology': 'Psychology',
    'Educational Psychology': 'Psychology'
}
```

```python
# Apply the mapping to the 'Major' column
df_2023['what_degree_program_are_you_currently_enrolled_in?'] = df_2023['what_degree_program_are_you_currently_enrolled_in?'].replace(major_mapping)
```

```python
# Verify the changes
df_2023['what_degree_program_are_you_currently_enrolled_in?'].unique()
```

```python
# Clean and standardize 'gender' column
gender_mapping = {
    'male': 'Male',
    'female': 'Female',
    'M': 'Male',
    'F': 'Female',
   
}

```

```python
# Apply the mapping to the 'gender' column
df_2023['gender'] = df_2023['gender'].replace(gender_mapping)
```
